## 1. Data Collection:

Below are listed all the libraries that will be imported to perform the data analysis tasks. The Dataset utilized for this analysis was from Kaggle, it is an online resource that contains public datasets and is free for everyone to choose from.

```python
# Importing Libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches
import seaborn as sns
from IPython.display import display
```

```python
# Load Data and adjust the amount of colums to display all
pd.set_option('display.max_columns', None)
d = pd.read_csv('all_seasons.csv')
d = d.drop(["Unnamed: 0"], axis = 1)
```

## 2. Data Cleaning
The NBA Dataset required its columns to be renamed in order for it to be accurate and easy to understand. The dataset required to be readjusted so the correct season was obtained for the analysis.

```python
# Code Cleaning 

# Rename Columns
d.columns = ["Player Name", "Team", "Age", "Height", "Weight", "College", "Country", "Draft Year", "Draft Round", "Draft No", "Games Played", "Pts", "Reb", "Ast", "Net Rating", "O-Reb pct", "D-Reb pct", "Usage pct", "FG pct", "Ast pct", "Season" ]

# Obtain the NBA season of 2022 - 2023
specific_szn = ['2022-23']
og_szn_2017to2023 = d.loc[d['Season'].isin(specific_szn)]
d1 = og_szn_2017to2023

# Readjust index after getting correct season
d1 = d1.reset_index()
d1 = d1.drop(["index"], axis = 1)
d1
```

```python
# Check to see if there are any NaN values (None were located)
d1[d1.Height.isna()]
```

### 2.1 Distributions

First, I will go over the distributions within the Dataset. The Figures below demonstrate the distribution of height amongst NBA players. From looking at both Figures most of the players are 200 cm tall (2 metres).

### 2.2 Data Conversion Numeric to Categoric for Height:
In order to correctly analyze and display the data. It was necessary to convert some data columns that were numeric into categorical datasets. This would help assist in finding the best-performing players by height. 

```python

# First we will divide the Height into 4 groups, using intervals of 12 centimetres 

# Intervals:
# 175 cm --- 187 cm  ---- Referred : Short
# 187 cm --- 199 cm  ---- Referred : Medium
# 199 cm --- 211 cm  ---- Referred : Tall
# 211 cm --- 224 cm  ---- Referred : X-Tall (Added 1 cm because of max height being 223.45)

copyd1 = d1

# This line of code helped create a new column within my exisiting dataset, the bins help distribute the values into its new categories such as, Short, Medium etc.
copyd1['Height group'] = pd.cut(x = copyd1['Height'], bins = [175, 187, 199, 211, 224], labels = ['Short', 'Medium', 'Tall', 'X-Tall'])

d2 = copyd1
display(d2)

 
```

Below represents the count of players which belong in the new height categories:

```python
hgroup_sum = copyd1['Height group']
print(hgroup_sum.value_counts())
```

### 2.3 Data Conversion Numeric to Categoric for Points, Rebounds, Assists and Games Played:
I will repeat the same method of converting the pre-existing numeric data into new categorical values. The numeric data that I will be converting into categorical data are Points(Pts), Rebounds(Reb), Assists(Ast) and Games Played. 

A summary will be printed after each conversion from numeric to categoric data to display the totals of each new group.

```python
# Dividing Points into new groups

# Intervals:
# 0 pts --- 5 pts    
# 5 pts --- 10 pts    
# 10 pts --- 15 pts   
# 15 pts --- 20 pts   
# 20 pts --- 25 pts   
# 25 pts --- 30 pts   
# 30 pts --- 35 pts   

# Operation from Numeric to Categoric Data
copyd1['Point group'] = pd.cut(x = copyd1['Pts'], bins = [-1, 5, 10, 15, 20, 25, 30, 35], labels = ['0 - 5 pts', '5 - 10 pts', '10 - 15 pts', '15 - 20 pts', '20 - 25 pts', '25 - 30 pts', '30 - 35 pts'])

p_group_sum = copyd1['Point group']
print(p_group_sum.value_counts())
```

```python
# Dividing Rebounds into new groups

# Intervals:
# 0 reb --- 3 reb  
# 3 reb --- 6 reb  
# 6 reb --- 9 reb   
# 9 reb --- 13 reb  

# Operation from Numeric to Categoric Data
copyd1['Rebound group'] = pd.cut(x = copyd1['Reb'], bins = [-1, 3, 6, 9, 13], labels = ['0 - 3 reb', '3 - 6 reb', '6 - 9 reb', '9 - 13 reb'])

r_group_sum = copyd1['Rebound group']
r_group_sum.value_counts()
```

```python
# Dividing Assists into new groups

# Intervals:
# First Interval: 0 ast --- 2 ast  
# Second Interval: 2 ast --- 4 ast   
# Third Interval: 4 ast --- 6 ast   
# Fouth Interval: 6 ast --- 8 ast  
# Fifth Interval: 8 ast --- 11 ast

# Operation from Numeric to Categoric Data
copyd1['Assist group'] = pd.cut(x = copyd1['Ast'], bins = [-1, 2, 4, 6, 8, 11], labels = ['0 - 2 ast', '2 - 4 ast', '4 - 6 ast', '6 - 8 ast', '8 - 11 ast'])

a_group_sum = copyd1['Assist group']
a_group_sum.value_counts()
```

```python
# Dividing Games Played into new groups

# Intervals:
# 0 ast --- 2 ast   
# 2 ast --- 4 ast  
# 4 ast --- 6 ast   
# 6 ast --- 8 ast   
# 8 ast --- 11 ast  

# Operation from Numeric to Categoric Data
copyd1['Games group'] = pd.cut(x = copyd1['Games Played'], bins = [-1, 20, 40, 60, 90], labels = ['0 - 20 games', '20 - 40 games', '40 - 60 games', '60 - 83 games'])

# Listed in reference for aid ^^^^
g_group_sum = copyd1['Games group']
print(g_group_sum.value_counts())

e1 = copyd1
```

Below Displays the updated Dataset that includes the new categorical columns at the end.

```python
e1 = copyd1
display(e1)
```
